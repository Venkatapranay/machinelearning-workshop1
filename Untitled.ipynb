## ELBOW method

```python
list1=[] # empty list
for k in range (1,10): # k=0,1,2,3..10
    kmeans2=KMeans(n_clusters=k)
    kmeans2.fit(x) #training
    list1.append(kmeans2.inertia_)   # KMeans inertia measures how well clusters are formed
```

```python
plt.plot(range(1,10),list1)
plt.show()
```

```python
last bent in graph is at 3 so total classifiers are 3 so choose n_clusters=3
```

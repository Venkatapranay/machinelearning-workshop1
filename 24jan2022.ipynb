```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(10,30,50)
y=np.sin(x)
plt.plot(x,y)
plt.plot(x,np.cos(x),color='green')
```

```python
help(plt.plot)
```

```python
plt.plot(x,y,marker='H',ms=10,mfc='white')
plt.plot(x,np.cos(x),color='green',marker='*',ms=20,mfc='red')
```

```python
#scatter plot
x=np.random.randint(100,1000,100)
y=np.random.randint(100,2000,100)
plt.scatter(x,y,color='red',marker='*')
```

```python
#bar plot
x=['19nov','19dec','20feb','20aug','21jan','21oct']
y=[10000,10500,5000,4000,8200,9015]
plt.bar(x,y,color='green')
```

```python
plt.bar(x,y,color=['green','c','m','pink','b','grey'])
```

```python
apps=['amazon','youtube','quora','ludo']
users=[10000,100000,50000,20000]
plt.pie(users,labels=apps,colors=['r','m','g','pink'],autopct='%1.2f%%')
plt.show()
```

```python
apps=['amazon','youtube','quora','ludo']
users=[10000,100000,50000,20000]
plt.pie(users,labels=apps,colors=['r','m','g','pink'],autopct='%1.2f%%',explode=(0,0,0.3,0.5))
plt.show()
```

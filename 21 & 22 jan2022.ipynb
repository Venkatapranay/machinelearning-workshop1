**replace null values**
- fillna()

- if data column is numerical replace with mean( )/median()/0

- if data column is object replace with most frequent value

```python
df=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Global%20Dataset/Market_Fact.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Product_Base_Margin'].mean()
```

```python
df['Product_Base_Margin'].fillna(df['Product_Base_Margin'].mean(),inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
import numpy as np
import pandas as pd
```

```python
dict1={'students':['lingareddy','uday','pranay','pranay',np.NaN],
        'year':[2,np.NaN,3,4,1],'branch':[np.NaN,'cse','eee','it','mec'],'specialized':['ncc','jee','sports',np.NaN,'ncc']}
data2=pd.DataFrame(dict1)
data2
```

```python
# dataframe['year'].fillna(method=ffill)
```

- replace all null of 'students' column with ffill

- replace all null of 'year' with 0

- replace all null of 'branch with bfill

- replace all null of 'specialized' with most frequent value

```python
data2['students'].fillna(method='ffill',inplace=True)
```

```python
data2['branch'].fillna(method='bfill',inplace=True)
```

```python
data2['year'].fillna(0,inplace=True)
```

```python
data2['specialized'].value_counts()
```

```python
data2['specialized'].fillna('ncc',inplace=True)
```
